Places you should take a tumble down the mandelbrot:

```python
import mandybrot as mandy
```

Settings:

```python
WIDTH = 1920
HEIGHT = 1080
MAX_ITERS = 100
SCALE = 1e-7
```

Locations:

```python
real = 0.42884
imag = -0.231345

data = mandy.sample.area(real, imag, WIDTH, HEIGHT, SCALE, MAX_ITERS)
cols = mandy.colour.image(data, MAX_ITERS, mandy.colour.jet_map)
mandy.colour.encode(cols)
```
